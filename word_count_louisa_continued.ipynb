```python
pp = open("pride_and_prejudice.txt", encoding='UTF8')
# encoding='UTF8' has solved the error with the code 'cp949'
di = dict()
for lin in pp:
    lin = lin.rstrip()
    wds = lin.split()
    for w in wds:
        di[w] = di.get(w,0) + 1
tmp = list()
for k,v in di.items() :
    newt = (v,k)
    tmp.append(newt)
tmp = sorted(tmp, reverse=True)
for v,k in tmp[:15] :
    print("pride and prejudice: ", k,v)

emma = open("emma.txt", encoding='UTF8')
# encoding='UTF8' has solved the error with the code 'cp949'
di = dict()
for lin in emma:
    lin = lin.rstrip()
    wds = lin.split()
    for w in wds:
        di[w] = di.get(w,0) + 1
tmp = list()
for k,v in di.items() :
    newt = (v,k)
    tmp.append(newt)
tmp = sorted(tmp, reverse=True)
for v,k in tmp[:15] :
    print("emma: ", k,v)

ss = open("sense_and_sensibility.txt", encoding='UTF8')
# encoding='UTF8' has solved the error with the code 'cp949'
di = dict()
for lin in ss:
    lin = lin.rstrip()
    wds = lin.split()
    for w in wds:
        di[w] = di.get(w,0) + 1
tmp = list()
for k,v in di.items() :
    newt = (v,k)
    tmp.append(newt)
tmp = sorted(tmp, reverse=True)
for v,k in tmp[:15] :
    print("sense and sensibility: ", k,v)

na = open("northanger_abbey.txt", encoding='UTF8')
# encoding='UTF8' has solved the error with the code 'cp949'
di = dict()
for lin in na:
    lin = lin.rstrip()
    wds = lin.split()
    for w in wds:
        di[w] = di.get(w,0) + 1
tmp = list()
for k,v in di.items() :
    newt = (v,k)
    tmp.append(newt)
tmp = sorted(tmp, reverse=True)
for v,k in tmp[:15] :
    print("northanger abbey: ", k,v)

ls = open("lady_susan.txt", encoding='UTF8')
# encoding='UTF8' has solved the error with the code 'cp949'
di = dict()
for lin in ls:
    lin = lin.rstrip()
    wds = lin.split()
    for w in wds:
        di[w] = di.get(w,0) + 1
tmp = list()
for k,v in di.items() :
    newt = (v,k)
    tmp.append(newt)
tmp = sorted(tmp, reverse=True)
for v,k in tmp[:15] :
    print("lady susan: ", k,v)

mp = open("mansfield_park.txt", encoding='UTF8')
# encoding='UTF8' has solved the error with the code 'cp949'
di = dict()
for lin in mp:
    lin = lin.rstrip()
    wds = lin.split()
    for w in wds:
        di[w] = di.get(w,0) + 1
tmp = list()
for k,v in di.items() :
    newt = (v,k)
    tmp.append(newt)
tmp = sorted(tmp, reverse=True)
for v,k in tmp[:15] :
    print("mansfield park: ", k,v)

wh = open("wuthering_heights.txt", encoding='UTF8')
# encoding='UTF8' has solved the error with the code 'cp949'
di = dict()
for lin in wh:
    lin = lin.rstrip()
    wds = lin.split()
    for w in wds:
        di[w] = di.get(w,0) + 1
tmp = list()
for k,v in di.items() :
    newt = (v,k)
    tmp.append(newt)
tmp = sorted(tmp, reverse=True)
for v,k in tmp[:15] :
    print("wuthering heights by Bronte: ", k,v)
```
