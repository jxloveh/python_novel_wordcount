```python
lw = open("little women.txt")
di = dict()
for lin in lw:
    lin = lin.rstrip()
    wds = lin.split()
    for w in wds:
        di[w] = di.get(w,0) + 1
tmp = list()
for k,v in di.items() :
    newt = (v,k)
    tmp.append(newt)
tmp = sorted(tmp, reverse=True)
for v,k in tmp[:10] :
    print("Little Women: ", k,v)

jos = open("jos_boys.txt", encoding='UTF8')
# encoding='UTF8' has solved the error with the code 'cp949'
di = dict()
for lin in jos:
    lin = lin.rstrip()
    wds = lin.split()
    for w in wds:
        di[w] = di.get(w,0) + 1
tmp = list()
for k,v in di.items() :
    newt = (v,k)
    tmp.append(newt)
tmp = sorted(tmp, reverse=True)
for v,k in tmp[:10]:
    print("Jo's Boys: ", k,v)

ec = open("eight_cousins.txt", encoding='UTF8')
# encoding='UTF8' has solved the error with the code 'cp949'
di = dict()
for lin in ec:
    lin = lin.rstrip()
    wds = lin.split()
    for w in wds:
        di[w] = di.get(w,0) + 1
tmp = list()
for k,v in di.items() :
    newt = (v,k)
    tmp.append(newt)
tmp = sorted(tmp, reverse=True)
for v,k in tmp[:10]:
    print("Eight Cousins: ", k,v)

rib = open("rose_in_blooms.txt", encoding='UTF8')
di = dict()
for lin in rib:
    lin = lin.rstrip()
    wds = lin.split()
    for w in wds:
        di[w] = di.get(w,0) + 1
tmp = list()
for k,v in di.items() :
    newt = (v,k)
    tmp.append(newt)
tmp = sorted(tmp, reverse=True)
for v,k in tmp[:10]:
    print("Rose in Blooms: ", k,v)

hs = open("hospital_sketches.txt", encoding='UTF8')
# encoding='UTF8' has solved the error with the code 'cp949'
di = dict()
for lin in hs:
    lin = lin.rstrip()
    wds = lin.split()
    for w in wds:
        di[w] = di.get(w,0) + 1
tmp = list()
for k,v in di.items() :
    newt = (v,k)
    tmp.append(newt)
tmp = sorted(tmp, reverse=True)
for v,k in tmp[:10]:
    print("Hospital Sketches: ", k,v)

bam = open("behind_a_mask.txt", encoding='UTF8')
# encoding='UTF8' has solved the error with the code 'cp949'
di = dict()
for lin in bam:
    lin = lin.rstrip()
    wds = lin.split()
    for w in wds:
        di[w] = di.get(w,0) + 1
tmp = list()
for k,v in di.items() :
    newt = (v,k)
    tmp.append(newt)
tmp = sorted(tmp, reverse=True)
for v,k in tmp[:10]:
    print("Behind a Mask; Or a Woman's Power: ", k,v)

md = open("moods.txt")
di = dict()
for lin in md:
    lin = lin.rstrip()
    wds = lin.split()
    for w in wds:
        di[w] = di.get(w,0) + 1
tmp = list()
for k,v in di.items() :
    newt = (v,k)
    tmp.append(newt)
tmp = sorted(tmp, reverse=True)
for v,k in tmp[:10]:
    print("Moods: ", k,v)
```
